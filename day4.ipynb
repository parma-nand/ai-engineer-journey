```python
categories = ["A", "B", "C"]
values = [5, 7, 3]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Values")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40]

plt.hist(data, bins=4)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
